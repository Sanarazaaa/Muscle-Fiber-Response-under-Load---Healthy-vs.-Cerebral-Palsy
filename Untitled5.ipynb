```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for muscle fiber mechanics under healthy and cerebral palsy conditions
length = 1.0  # Original length of the muscle fiber (m)
cross_sectional_area = 0.01  # Cross-sectional area (m^2)
youngs_modulus_healthy = 1e6  # Young's modulus for healthy muscle (Pa)
youngs_modulus_cp = 3e5  # Reduced Young's modulus (Pa) for cerebral palsy
force = 1000  # Applied force (N)
num_elements = 10  # Number of finite elements

# Length of each element
element_length = length / num_elements

# Calculate stiffness matrix for each element
def element_stiffness(E, A, L):
    return E * A / L

# Generate the global stiffness matrix (for a 1D linear system)
def assemble_global_stiffness_matrix(E, A, num_elements, element_length):
    global_matrix = np.zeros((num_elements + 1, num_elements + 1))

    # Element stiffness for each element
    k = element_stiffness(E, A, element_length)

    for i in range(num_elements):
        global_matrix[i, i] += k
        global_matrix[i, i + 1] -= k
        global_matrix[i + 1, i] -= k
        global_matrix[i + 1, i + 1] += k

    return global_matrix

# Calculate the force vector (assuming force is applied at the last node)
def force_vector(force, num_elements):
    f = np.zeros(num_elements + 1)
    f[-1] = force  # Applying force at the last node
    return f

# Apply boundary conditions (assuming fixed displacement at the first node)
def apply_boundary_conditions(global_matrix, force_vector):
    # Fix the first node (zero displacement)
    global_matrix[0, :] = 0
    global_matrix[0, 0] = 1
    force_vector[0] = 0
    return global_matrix, force_vector

# Solve for displacements
def solve_displacements(global_matrix, force_vector):
    return np.linalg.solve(global_matrix, force_vector)

# For healthy muscle
global_matrix_healthy = assemble_global_stiffness_matrix(youngs_modulus_healthy, cross_sectional_area, num_elements, element_length)
force_vector_healthy = force_vector(force, num_elements)
global_matrix_healthy, force_vector_healthy = apply_boundary_conditions(global_matrix_healthy, force_vector_healthy)

displacements_healthy = solve_displacements(global_matrix_healthy, force_vector_healthy)

# For cerebral palsy muscle
global_matrix_cp = assemble_global_stiffness_matrix(youngs_modulus_cp, cross_sectional_area, num_elements, element_length)
force_vector_cp = force_vector(force, num_elements)
global_matrix_cp, force_vector_cp = apply_boundary_conditions(global_matrix_cp, force_vector_cp)

displacements_cp = solve_displacements(global_matrix_cp, force_vector_cp)

# Plotting the deformed shapes of the muscle fibers
x = np.linspace(0, length, num_elements + 1)
y_healthy = displacements_healthy
y_cp = displacements_cp

plt.figure(figsize=(8, 4))
plt.plot(x, y_healthy, label='Deformed Length (Healthy)', color='green')
plt.plot(x, y_cp, label='Deformed Length (Cerebral Palsy)', color='red')
plt.title('Muscle Fiber Response under Load - FEA (Healthy vs Cerebral Palsy)')
plt.xlabel('Position along the Fiber (m)')
plt.ylabel('Displacement (m)')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', lw=0.5)
plt.show()

# Display results for comparison
print("FEA Simulation - Healthy vs Cerebral Palsy")
print(f"Original Length: {length} m")
print(f"Applied Force: {force} N")
print(f"Displacement at nodes (Healthy): {displacements_healthy}")
print(f"Displacement at nodes (Cerebral Palsy): {displacements_cp}")
```
